A human and a rock are both made of the same types of atoms and follow the same laws of physics. However, the arrangements of atoms that make up animate matter tend to be large and structurally complex compared to the simple orderly structures in inanimate matter, and are hence called macromolecules (or large molecules). The complexity of biological macromolecules is such that there are more possible protein structures than there are protons in the universe.

This quiz will serve as an introduction to the major players in the central dogma of biology: the remarkable process of how information turns into biological order.

ince dead matter and living matter are often made of the same stuff, the accepted characteristics of life are largely empirical. Living matter tends to exist far from thermal equilibrium, with well-regulated processes of growth which consume inanimate matter and energy from the environment, reproduce to create similar offspring, and eventually die.

These features of life are enabled by the unique self-perpetuating processes of biology which regulate and animate a soup of sugar, salt, and oil which would otherwise rapidly settle into a silent equilibrium.

If we take any living organism from a newborn mayfly to a 300-year-old oak tree and inspect its constituent macromolecules, we will find four common classes: nucleic acids, proteins, carbohydrates (sugar-containing molecules), and lipids (fat-containing molecules). Each of these classes serves distinct roles in the living cell.

Nucleic acids such as DNA and RNA serve as operating instructions that guide the construction of every different type of macromolecule including themselves. Proteins serve many roles due to their incredible chemical diversity, including accelerating chemical reactions, quickly transporting smaller molecules, and serving as a structural element of a cell. Lipids form barriers between the cell and its environment and compartmentalize the cell into compartments called organelles. Carbohydrates are used for energy storage, and as the building blocks in rigid structures called cell walls.

How might life manage to synthesize such a diversity of structures?

    Repeating the same few chemical reactions over and over again

The diversity of structures is enabled by combinatorial assembly of a small set of common building blocks. The biological machinery in a cell is capable of performing a relatively restricted repertoire of chemical reactions for the disassembly and assembly of certain building blocks. The ability to disassemble allows the transformation of biomolecules scrounged from the environment into common building blocks, which are then combinatorially assembled into new structures.

This combinatorial assembly repeats the same few chemical reactions over and over again with different building blocks, and leads to the diversity of biomolecules with relatively simple chemistry. DNA is a linear strand assembled by repeating one reaction with a different order of component nucleotides, and proteins are similarly constructed with repeated assembly of amino acids in a different order. The space of permutations of relatively few building blocks becomes massive very rapidly with the increasing length of the macromolecule.

DNA contains the blueprints for every protein and other structure in the cell. One blueprint at a time, DNA is transcribed into RNA which is delivered where and when it's required to be translated into a protein. Proteins then perform a relatively small number of chemical reactions which are required for the assembly of inanimate building blocks collected from the environment into more DNA, RNA, proteins, and other biological structures. The rest of the behavior, including what raw materials are consumed, how energy is produced, and how reproduction occurs, is called the phenotype.

The genes responsible for encoding this self-perpetuating molecular assembly are among the <b>immortal</b> genes which were explored in the last chapter and found to possess incredible protection from significant mutations over evolutionary timescales. These DNA sequences are central to the fundamental and universal processes of life and thus have persisted in all domains of life including archaea, bacteria, plants, and animals over the past two billion years with minimal structural or functional variation.

We've seen that DNA is always a double-stranded rigid chain. RNA, on the other hand, can be a lot more flexible. It's made up of a string of nucleotides just like DNA, but is often found in its single-stranded form. <b>This flexibility also significantly decreases RNA's stability by a million-fold compared to DNA.</b>

The short-lived nature of RNA is not a bug but a feature, which is even more pronounced for long messenger RNA (mRNA) chains in an organism, where strands usually stay intact for less than 5 minutes. In this time, RNA can't spread to different parts of the cell or organism without being degraded. The fleeting existence of RNA is ideal for its role as a messenger from the genome to the protein assembly line where translation occurs. 

<b>Why is a short-lived molecule well-suited to the role of messenger?</b>
Explanation

Correct answer: More spatial and temporal control in protein production

The synthesis of a single RNA transcript will lead to protein synthesis only for a short pulse of time. If the circumstances of the cell or its environment change, protein synthesis will not need to be "turned off," as the messenger RNA rapidly degrades. This simple consequence of a shorter molecular half-life enables the ``turn-on" regulation of gene expression and allows the same genome to code for heart cells and hair cells.

The spatial and temporal resolution of RNA transcription is particularly useful during the development stages of advanced organisms like flies. As a fly embryo proceeds through the various stages of development from a single egg to a full-grown fly, its cells express RNAs in a specific spatial and temporal pattern that aids in the formation of eyes and wings at the right places, and bilateral symmetry.

It is this same spatiotemporal gene expression that leads to patterns in organisms including a tiger's stripes. If RNA was stable and could diffuse all around an organism, it wouldn't be possible to create a pattern like stripes, since the RNA which codes for the different colors could spread throughout.

But RNA isn't always a messenger. RNA is much less picky about having a perfectly matched and complementary double helix compared to DNA. In fact, RNA will usually form imperfectly matched double-stranded structures with itself, resulting in folded structures much like proteins, but based only on the two A:U and C:G

base-pairs (with rare exceptions).

Many RNA sequences are never translated, and instead these stable RNA folds serve a structural, regulatory, or catalytic function in the cell. Arguably the most complex molecular machine known is the ribosome, a giant folded RNA and protein structure which speaks both of the great languages of biology and is responsible for translating RNA transcripts into protein sequences. We'll explore how the ribosome works by walking through several of the most famous experiments of molecular biology in an upcoming quiz.

Proteins are made up of amino acids, and are completely different molecular chains compared to the nucleotides of DNA and RNA. Still, their structure begins as a linear chain about 100 amino acids long, which is encoded by one gene. Proteins rapidly fold into defined secondary structures like helices and sheets, which we briefly explored in the last chapter. These structures are then arranged to form a three-dimensional fold called the native structure which is held together by a large number of different bonds. A protein can only function once it's folded into this structure.

Evolution has lead to an amazing variety of protein structures and functions, many of which we have yet to fully understand.

Protein sequences tend to be conserved even across distantly related species and organisms. This suggests that a protein which performs a given task must preserve most of its structure to still be useful. An ancient ancestral protein sequence moves through sequence space as it mutates and evolves, but selection pressure forces the sequence to remain highly conserved and clustered in the same region of sequence space.
