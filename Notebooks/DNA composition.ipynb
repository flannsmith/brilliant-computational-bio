Chemical analysis of DNA quickly identified it consisted of four component nucleotides (A,T,C,G)

These components were not recognized as letters which encoded information for quite some time. Phoebus Levene, the man who first discovered the four nucleotides claimed that DNA was made up of long repetitions of the four nucleotides stacked on one another in an invariant fashion. He called this theory the tetra-nucleotide hypothesis (or TNH), but others referred to it as the "pancake model." Thanks to Levene's reputation, this theory of DNA structure went largely unchallenged. In search of life's information storage molecule, biologists busied themselves developing models based on chains of sugar (like starch) and protein.

Following some suggestive experiments of his time, Erwin Chargaff doubted the TNH and sought to disprove it by directly measuring the relative levels of the four nucleotides to see whether they were truly equal.

Phoebus Levene was one of the premier organic chemists of his day, and even with relatively primitive equipment and techniques, he was able to deduce correctly that a nucleotide is composed of three molecules bound together: 
        * a deoxyribose sugar, 
        * an inorganic phosphate, 
        * and a nitrogenous base. 
 Furthermore he found that there are four different bases which distinguish one nucleotide from another: two larger bases called adenine and guanine and two smaller bases called thymine and cytosine

Levene and others had found that certain cellular structures (called chromosomes for their deep coloring under a microscope) were rich in both proteins and DNA. For nearly a century biologists had observed that chromosomes are duplicated and lined up during cellular reproduction, with an equal amount going to each daughter cell, and that these structures are the carriers of hereditary information.

Due to the high DNA content in chromosomes, many scientists considered DNA a likely hereditary molecule. But once the highly reputed Levene conclusively discovered the molecular structure of the nucleotide and proposed the TNH, the scientific community was faced with a major problem: the molecule which carries hereditary information must contain a large amount of data to code for all the rich variety of organisms.

The invariant tetranucleotide hypothesis proposed by Levene states specifically that there is only one building block of DNA sequences. An endless repetition of an invariant tetranucleotide cannot encode information of any type, that is, any single word repeated endlessly can never on its own convey semantic information.

To most scientists of the day, DNA was seen as a "dumb" structural molecule, not unlike the endless sugar chains which make up starch in plants or exoskeletons in insects. Proteins were known to consist of twenty possible building blocks called amino acids, allowing each amino acid to encode 
log2(20) = 4.3
bits of information. This high information density and the presence of protein in chromosomes convinced most that proteins were the information storage medium of heredity.

Erwin Chargaff was unconvinced. He questioned the TNH and the accuracy of Levene's experiments, and sought to accurately measure the relative abundance of the DNA bases with new technologies. We will perform a similar experiment in silico.

Chargaff carefully collected samples of DNA from many different organisms and broke them down to their constituent nucleotide building blocks. Each organism's genome contains millions to billions of nucleotides which he would analyze using newly invented chromatography and spectroscopy techniques. For our in silico analysis, we are lucky that high-throughput DNA sequencing has generated enormous datasets with millions or billions of nucleotides that scientists have made available free online. Computational techniques for building and analyzing such datasets will be the focus of several chapters of this course.

A genetic sequence can be represented straightforwardly in Python as a string:

```python
def thymine_counter(sequence):
    thymines = 0
    for base in sequence:
        if base is "T":
            thymines = thymines + 1
    return thymines
```

```python
def composition(string):
    # Initialize all the counters to zero
    (thymines, cytosines, adenosines, guanines) = (0, 0, 0, 0)
    for base in string:
        if base is "A":
            adenosines = adenosines + 1
        elif base is "T":
            thymines = thymines + 1
        elif base is "C":
            cytosines = cytosines + 1
        elif base is "G":
            guanines = guanines + 1
    return {"adenosine" : adenosines, "thymine" : thymines,
            "cytosine" : cytosines, "guanine" : guanines}
```
