In fact, biological engineering has advanced far enough recently that without a genetic sequence ever leaving a computer, we can follow its instructions and construct all the proteins and structures that make up a cell in a laboratory. Suppose a laboratory were to construct all the proteins as instructed by the genome and put them together into a whole new synthetic cell.

Would this synthetic cell be fully functional and able to self-replicate (make a functioning copy of itself)?


No, even a complete cell is unable to self-replicate without its own copy of the genome 



Cells and bioengineering laboratories are able to manufacture and assemble a wide variety of proteins and cellular structure that could make up another cell. They can even combine them all together. But the newly assembled cell would not be able to self-replicate without a copy of its genome.

The cell itself is able to do many things, but without a genome it can never self-replicate. Additionally, if a cell doesn't have a copy of its genome, it won't be able to respond to different environments and stimuli to perform different tasks. It's not only necessary to build another cell, but the information in the genome must also be copied and transferred to the new cell.


On its own, a cell is just a chemically complex but inert droplet of proteins and other molecular machines. The information contained in the genome animates the cell and is a crucial ingredient for a living organism. Without a genome, cells cannot react to their environment or construct a copy of themselves.

When biologists perform cloning, they often remove the genome from a cell, leaving behind the proteins and molecular machines capable of following genetic instructions. Then they add in a new DNA genome and wait. The resulting cell and any of its offspring is completely transformed: they don't share any similarity with the original cell since it only followed the new genetic instructions that were added.

But at the same time, without the molecular machinery of a cell, the genome is little more than a very efficient molecular memory storage system. A real-world example of an isolated genome is a virus. Viruses are essentially tiny-packaged genomes containing the instructions to make more of themselves.

Why do biologists not consider viruses alive?

 Correct answer: Viruses cannot make copies of themselves without the help of other cells

Although viruses are capable of proliferating and making copies of themselves, they are not considered "alive" in the traditional sense. This is because, on their own, viruses are not animated (i.e. they do not perform any tasks or respond to their environment), and they cannot self-replicate.

Viruses depend on infected host cells to provide most of the biological machinery that makes their replication possible.

This is the bare-minimum requirement that scientists use to define a life form:

    Given a crucial dependence on the materials and energy it extracts from the environment, the cell is a self-replicating entity.


A cell is not simply a squishy bag of parts that can perform a set of biological processes—it must have the ability to self-replicate by making a copy of its components and also copy its genome into a new cell.

The information required to make a cell is organized in a modular fashion. This finding has enabled the field of recombinant DNA, where modules encoding certain functions can be moved from one organism to another. One of the most famous recombinant DNA technologies is called GFP (or green fluorescent protein). This gene was first found in a jellyfish, but today it's been moved to thousands of different species, including mice. By introducing mutations in the GFP gene, scientists have successfully color shifted GFP to produce red, blue, yellow, and orange fluorescent proteins.

Completely deleting the genes for certain proteins is an effective (if barbaric) method for teasing out the function of each gene. Biologists call this type of approach knocking-out a cell's genes one at a time in order to determine their functionality. Knock-out cells are missing a gene and unable to perform a given function, such as biofilm formation (dispersin) or locomotion (flagellin).

This type of study yields several important pieces of information about the cell's instructions: The specific location of each gene within the genome, the gene sequence that leads to a fully functional cell, and the function linked to each gene.

Sometimes it's not so simple. The gene which codes for glucose dehydrogenase is essential for metabolizing sugar for energy, but a knock-out of this gene doesn't change the cell's metabolism at all, they still happily gain energy from glucose (sugar) in their environment. What might explain this?

<b> Correct answer: Important functions often have redundant genes.</b>

We'll notice over and over again in this course that the solutions life has found to deal with the world around it and the limitations of physical law are incredibly resourceful. Mutations happen, and in order to maximize the likelihood of an organism surviving, the most essential cellular functions are encoded redundantly in several (or many!) different genes.

One example of this is the sugar metabolism performed by glucose dehydrogenase. This is an essential protein required to extract energy from the cell's most common nutrient (sugar), and so knocking out a single gene will never completely destroy it's metabolic ability. This function is encoded in several other genes scattered around the genome so that a single knock out or mutation can't lead to an evolutionary dead end for an organism.

Each of the N regions of the genome can be included or not, meaning there are configurations for each gene. Given N

possible genome configurations.

In real genomes, interdependencies between different proteins and functions drastically constrain the set of alterations we can make while still resulting in a living cell. A knock-out or rearrangement study of a cell's genes will likely yield no functional cells for a large fraction of the possible permutations. In many cases, genomes are actually long circular loops of DNA, rather than linear chains. Biological constraints on the roles of genes often limit their arrangements to circular clusters with a consistent connectivity. Thus any circular rearrangement of genes can encode functional cells.

For example, the replication machinery in bacteria can only copy a DNA sequence at a maximum speed of about 1000 nucleotides per second, and in each 20 minute generation the 5x10^6 nucleotide genome needs to be copied in its entirety. How might this be possible?


### Explanation

<b>Correct answer: The cell must copy its genome in parallel in each generation</b>

Since a single set of replication machinery can only copy the whole genome in about

minutes, there must be several working in concert and parallelizing the replication process.

It turns out that bacteria under ideal growth conditions (with a 20 minute generation time) use nested replication forks that begin replicating the granddaughter and great-granddaughter cells' genomes while the daughter cells are still themselves engaged in replication.
