The mRNA for a given gene is built by unwinding the genome’s DNA double helix in the region of the gene, so as to expose its nucleotides. With the nucleotides accessible, each DNA nucleotide on the template strand is paired with its RNA complement one nucleotide at a time, unfurling the RNA in the 5' to 3' sense ordering from single nucleotide building blocks.

```python
Image('Img/transcription.png')
```

The base-pairing and assembly of a new RNA strand is coordinated by an amazing, solitary machine called <b>RNA polymerase</b>, so-named because it polymerizes (or links together) new RNA sequences. This step of the central dogma is known as transcription, as it involves making a transcript of one gene's blueprint from the genome.

What is the relationship between a gene’s coding sequence in the genome and its RNA encoding?

<b>Exact copy apart from the T -> U conversion</b>

We know from our previous adventures with DNA that the two strands in DNA are complementary to one another, following the base-pairing rules. So, in this example, the template strand must be complementary to the coding strand, and vice versa. 

We also know that the RNA transcript is constructed by RNA polymerase to be complementary to the template strand. Thus, since both the new RNA strand and the coding strand are complementary to the template, they must be the same sequence. <b>The only difference is that RNA is written with a U instead of a T.</b>

DNA is almost always double-stranded, with both a <b>coding</b> strand and a complementary <b>template</b> strand in the reverse direction. Usually only one of these strands encodes an mRNA, but there are two ends to start reading from, each of which would end up as a different RNA sequence. How can the RNA polymerase (or an enterprising biologist given a DNA molecule) know which strand to read and transcribe?

```python
Image('Img/seqeunce.png')
```

Biology has solved this problem by using a <b>promoter</b> sequence, a canonical example of which is the "TATA box," a short nucleotide sequence <b>TATAAAA</b> found on the coding strand just before the start of the gene. The RNA polymerase only starts transcribing when it encounters a TATA box, and as enterprising computational biologists, we'll use this signal to mine DNA sequences for possible genes.

### Make an mRNA transcript

To make an mRNA transcript from an unknown DNA string, we need to do several steps to make sure that we are transcribing the right sequence:

1. Generate a complementary DNA sequence just in case the gene is encoded on the complementary strand.

2. Reverse the direction of the complementary sequence to the 5' to 3' sense direction.

3. Find which strand has a TATA box promoter (the coding strand).

4. Perform a T -> U substitution

```python
def complement(Sequence):
    compseq = []
    ## Iterate through the sequence string, replacing each nucleotide with its complement
    ## according to the base-pairing rules
    for i in Sequence:
        if i == "T":
            compseq.append("A")
        elif i == "A":
            compseq.append("T")
        elif i == "G":
            compseq.append("C")
        elif i == "C":
            compseq.append("G")
    ## Return a combined string of the complement sequence.
    return ''.join(compseq)
```
