This is the process through which information stored in DNA is expressed, or transformed into RNA, proteins and other cellular structures. DNA and the shortlived RNA transcripts both speak the language of nucleotides, though there are subtle differences in their sequences and structure. Once the RNA is translated into proteins, the molecular language must change. This is carried out by an amazing molecular machine known as the ribosome, along with a dictionary of helper molecules known as tRNAs. 

DNA and RNA also share an incredibly important feature: they can only be assembled in one direction.
